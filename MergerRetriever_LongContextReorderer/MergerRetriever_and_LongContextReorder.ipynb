# Vector Embedding

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
bge_embeddings = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-large-en")
```

# Data Ingestion

```python
from langchain_community.vectorstores import Chroma
import chromadb
```

```python
import os
os.getcwd()
```

```python
CURRENT_DIR = os.path.dirname(os.path.abspath("."))
```

```python
CURRENT_DIR
```

```python
DB_DIR = os.path.join(CURRENT_DIR, "/content/db")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False,
)
```

```python
harrypotter_vectorstore = Chroma.from_documents(
    documents=text_harrypotter,
    embedding=bge_embeddings,
    collection_name="harrypotter",
    persist_directory="/store/harrypotter",
    collection_metadata={"hnsw:space": "cosine"},
)
```

```python
got_vectorstore = Chroma.from_documents(
    documents=text_got,
    embedding=bge_embeddings,
    collection_name="got",
    persist_directory="/store/got",
    collection_metadata={"hnsw:space": "cosine"},
)
```

```python
retriever_harrypotter = harrypotter_vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 5}
)

retriever_got = got_vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 5}
)
```

```python
from langchain_classic.retrievers import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_harrypotter, retriever_got])
```

```python
docs = lotr.invoke("Who is harry potter?")

for chunk in docs:
    print(chunk.page_content)
```

# Pipeline for LLM

```python
# ===== Document Transformers =====
from langchain_classic.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
    LongContextReorder
)

# ===== Document Compressor / Retriever =====
from langchain_classic.retrievers.document_compressors import DocumentCompressorPipeline
from langchain_classic.retrievers import ContextualCompressionRetriever
```

```python
from re import search
filter = EmbeddingsRedundantFilter(embeddings=bge_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr,search_kwargs={"k": 3, "include_metadata": True}
)
```

```python
docs = compression_retriever_reordered.invoke("What is ESOPs?")
print(len(docs))
print(docs[0].page_content)
```

# LLM Initialization

```python
!pip install llama-cpp-python
!pip install -q huggingface_hub
```

```python
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(
    repo_id="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
    filename="tinyllama-1.1b-chat-v1.0.Q2_K.gguf",
)
```

```python
from langchain_classic.llms import LlamaCpp

llm = LlamaCpp(
    model_path=model_path,
    streaming=True,
    max_tokens=512,
    temperature=0.7,
    n_ctx=2048,
    verbose=True,
)
```

```python
from langchain_classic.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # same as your old code
    retriever=compression_retriever_reordered,
    return_source_documents=True
)
```

```python
query = "Who is Jon Snow?"
result = qa(query)

print("Answer:\n", result['result'])
print("\nSource documents:\n")
for doc in result['source_documents']:
    print(doc.page_content[:300], "\n---\n")
```
